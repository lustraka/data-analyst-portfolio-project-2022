## Results

````python
def parse_print_string(s, delimiter='|'):
    '''Parse a text and print it with wrapping.'''

    if str(s) == 'nan':
        print('\t','nan')
    else:
        lines = s.split(delimiter)
        for line in lines:
            for txtline in textwrap.wrap(line.strip(), width=80):
                print('\t',txtline)

def print_column(df, name, delimiter):
    '''Print the content of the given column'''

    for idx,rec in df.iterrows():
        print('\n'+str(idx+1), rec['study_name'].title())
        print('-'*84)
        parse_print_string(rec[name], delimiter)

# Take only records from a training set
des_train = des.loc[des['your comment'] == 'train']


# Print selected column
print_column(des_train, 'research_question', '|')
````

````python
# Print selected column
print_column(des_train, 'data_extracted', ',')
````
